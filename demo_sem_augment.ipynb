```python
import skrub

import sempipes

dataset = skrub.datasets.fetch_credit_fraud()
raw_products = dataset.products.head(n=500)

products = skrub.var("products", raw_products)
products
```

```python
products_augmented = products.sem_augment(
    nl_prompt="Generate additional products from APPLE.",
    number_of_rows_to_generate=500,
    generate_via_code=True,
    name="augment_with_more_apple",
)
```

```python
products_augmented
```

```python
sempipes.inspect_generated_code(products_augmented, "augment_with_more_apple")
```
